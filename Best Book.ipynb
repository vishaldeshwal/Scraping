```python
import pandas as pd
import bs4
import selenium
import requests
from selenium import webdriver
```

```python
import os
os.getcwd()
```

```python
browser = webdriver.Chrome('chromedriver.exe')
browser.get('https://www.goodreads.com/choiceawards/best-books-2017')
```

```python
books = browser.find_elements_by_css_selector('div.category.clearFix')
print(len(books))
```

```python
df_products = pd.DataFrame(columns = ['category','book_name','url_cover_page','url_book_description'])
for book in books:
    category = book.find_element_by_css_selector('h4')
    category = category.get_attribute('innerHTML') #gets the text from line, if we use .text it will noit return some names
    book_name = book.find_element_by_css_selector('img.winner')
    book_name = book_name.get_attribute('alt')
    url_cover_page = book.find_element_by_css_selector('img.winner')
    url_cover_page = url_cover_page.get_attribute('src')
    url_book_description = book.find_element_by_css_selector('a')
    url_book_description = url_book_description.get_attribute('href')
    curr_row = {
       'category' : category,
       'book_name' : book_name,
       'url_cover_page' : url_cover_page,
       'url_book_description' : url_book_description
    }
    df_products = df_products.append(curr_row, ignore_index=True)
df_products
```

```python
for url in df_products['url_book_description'].head(3):
    browser.get(url)
    author = browser.find_element_by_css_selector('span[itemprop="name"]').get_attribute('innerHTML')
    desc = browser.find_element_by_css_selector('div.readable.stacked.gcaBookDescription').get_attribute('innerHTML')
    votes = browser.find_element_by_css_selector('span.greyText.gcaNumVotes').get_attribute('innerHTML')
    print(author,'|',desc,'|',votes)
```
